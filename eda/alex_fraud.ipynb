```python
def avg_price(tic_lst):
    res = 0
    numer = 0
    denom = .001
    for tic in tic_lst:
        # get our total quantity of tickets
        denom += tic['quantity_total']
        numer += tic['cost'] * tic['quantity_total']
    
    res = numer / denom
    return res
```

```python
my_columns['avg_price'] = my_columns['ticket_types'].apply(avg_price)
```
